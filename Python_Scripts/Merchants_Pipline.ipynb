```python
import pandas as pd
import mysql.connector as lol
import json
import collections
from sys import path
import pyodbc 



conn_mysql = lol.connect(user='root', password='P@ssw0rd',host='localhost',database='cowpay_staging')
conn_sql = pyodbc.connect('Driver={SQL Server};''Server=BI-SERVER2022;''Database=Staging_Migration;''Trusted_Connection=yes;')

cursor = conn_mysql.cursor()
data = pd.read_sql("SELECT * from cowpay_staging.users where is_merchant=1",conn_mysql)
my_data = data.to_dict()

for key, value in my_data.items():
    if key == "m_facebook":
        facebook = value
    if key == "m_twitter":
        twitter = value
    if key == "m_linked_in":
        linked = value
    if key == "m_instagram":
        instagram = value
    if key == "m_website":
        website = value

SocialLinks_dict = {}
for index, d in enumerate([facebook, twitter, linked, instagram, website]):
    for k, v in d.items():
        if index == 0: #facebook
            SocialLinks_dict.setdefault(k, []).append(json.dumps({
                "SocialUrl": v,
                "SocialType": {"id": "1", "Name": "Facebook", "Icon": "fab fa-facebook-f"}
            }))
        if index == 1: #twitter
            SocialLinks_dict.setdefault(k, []).append(json.dumps({
                "SocialUrl": v,
                "SocialType": {"id": "2", "Name": "Twitter", "Icon": "fab fa-twitter"}
            }))
        if index == 2:  # linked
            SocialLinks_dict.setdefault(k, []).append(json.dumps({
                "SocialUrl": v,
                "SocialType": {"id": "3", "Name": "LinkedIn","Icon":"fas fa-link"}
            }))
        if index == 3:  # instagram
            SocialLinks_dict.setdefault(k, []).append(json.dumps({
                "SocialUrl": v,
                "SocialType": {"id": "5", "Name": "m_instagram","Icon":"fas fa-instagram"}
            }))
        if index == 4:  # website
            SocialLinks_dict.setdefault(k, []).append(json.dumps({
                "SocialUrl": v,
                "SocialType": {"id": "4", "Name": "Website","Icon":""}
            }))

data['SocialLinks'] = SocialLinks_dict.values()
data['SocialLinks']


for key, value in my_data.items():
    if key == "m_tax_id_file":
        TaxIdImageUrls = value
    if key == "m_personal_id_file":
        PersonalIDImageUrls = value   
    if key == "m_commercial_registry_file":
        CommercialRegistrationImageUrls = value
    
  
        
AttachamentImages_dict = {}
for index, d in enumerate([TaxIdImageUrls, PersonalIDImageUrls, CommercialRegistrationImageUrls]):
    for k, v in d.items():
        if index == 0: #facebook
            AttachamentImages_dict.setdefault(k, []).append(json.dumps({
              "TaxIdImageUrls":v
              
            }))      
                                             
        if index == 1: #facebook
            AttachamentImages_dict.setdefault(k, []).append(json.dumps({
              "PersonalIDImageUrls":v
              
            }))
        if index ==2: #facebook
            AttachamentImages_dict.setdefault(k, []).append(json.dumps({
              "CommercialRegistrationImageUrls":v
              
            }))
                                             
data["AttachamentImages"]=AttachamentImages_dict.values()   
#print(data["AttachamentImages"])
data.to_csv('AI_JSON.csv')
```
